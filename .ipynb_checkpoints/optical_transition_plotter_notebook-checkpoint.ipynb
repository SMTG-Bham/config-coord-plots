```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.patches import FancyBboxPatch
%matplotlib inline
```

```python
def calculate_absorption_energy(band_gap, E_i_q, E_i_qplusone, q, qplusone):    
    print(f"Calculating absorption energy of {q} to {qplusone} transition:")
    abs_energy = band_gap - ((E_i_q - E_i_qplusone)/(qplusone - q))
    print(abs_energy)
    return abs_energy

def calculate_relax_excited_energy(E_i_qplusone, E_f_qplusone):
    print("Calculating energy of excited state relaxation (phonon process):")
    relax_excited_energy = E_i_qplusone - E_f_qplusone
    print(relax_excited_energy)
    return relax_excited_energy

def calculate_emission_energy(band_gap, E_f_qplusone, E_f_q, q, qplusone):  
    print(f"Calculation emission energy of {qplusone} to {q} transition:")
    em_energy = band_gap - ((E_f_qplusone - E_f_q)/(q - qplusone))
    print(em_energy)
    return em_energy

def calculate_relax_ground_state_energy(E_f_q, E_i_q):
    print("Calculating energy of ground state relaxation (phonon process):")
    relax_ground_state_energy = E_f_q - E_i_q
    print(relax_ground_state_energy)
    return relax_ground_state_energy

def acceptor_to_cbm_calculator(band_gap, q, qplusone, E_i_q, E_i_qplusone, E_f_qplusone, E_f_q):
    
    """
    Calculates full absorption and emission profile energies. Returns a list of energies.
    Refer to example diagram to further understand the process.
    You will need to have calculated the formation energies of each charge state in each defect geometry.
    
    
    \tband_gap = material band gap.
    \tq = initial charge state.
    \tqplusone = final charge state (after absorption).
    \tE_i_q = formation energy of defect in charge state q in initial equilibrium defect geometry. (before absorption)
    \tE_i_qplusone = formation energy of defect in charge state q+1 but in initial equilibrium geometry. (immediately after absorption)
    \tE_f_qplusone = formation energy of defect in charge state q+1 in final equilibrium geometry. (before emission)
    \tE_f_q = formation energy of defect in charge state q but in final equilibrium geometry. (immediately after emission)
    
    
    
    NOTE: see example plot for graphical description of energy levels.
    """
    print("Calculating optical transitions...")
    abs_energy = calculate_absorption_energy(band_gap, E_i_q, E_i_qplusone, q, qplusone)
    relax_1 = calculate_relax_excited_energy(E_i_qplusone, E_f_qplusone)
    em_energy = calculate_emission_energy(band_gap, E_f_qplusone, E_f_q, q, qplusone)
    relax_2 = calculate_relax_ground_state_energy(E_f_q, E_i_q)
    print("Finished.")
    return [abs_energy, relax_1, em_energy, relax_2]
    
```

```python
def optical_transition_plot(list_of_defects=["$A^{q}$", "$A^{q+1}$"],
                            list_of_energies=[["$E_{abs}$", "$E_{phonon}^{q+1}$", "$E_{em}$", "$E_{phonon}^{q}$"]],
                            example_plot=True,
                            example_show_variable_labels=False,
                            example_show_process_labels=False,
                            plot_title="optical_transition",
                            phonon_labels=False,
                            list_of_plotting_colours=["silver", "red", "blue"],
                            convert_abs_em_to_nm=False,
                            fmt="pdf"):
    """
    Constructs an optical transition plot for defects using the energies calculated from acceptor_to_cbm_calculator().
    
    list_of_defects = comma separated list of LaTeX formatted labels for the plot. 
    \tFor example, a neutral copper vacancy would be entered as "$V_{Cu}^{0}$. 
    \tDefault: ["$A^{q}$", "$A^{q+1}$"] for generating example plot.
    list_of_energies = list of outputs from acceptor_to_cbm_calculator() function.
    \tCan also enter a list of function calls.
    \tFor example, list_of_energies = [[w,x,y,z], [a,b,c,d]] where [w,x,y,z] is the output of one set of transitions,
    \tand [a,b,c,d] is the output of another set.
    \tIf only one set of energies, still insert as a list.
    \tDefault: [["$E_{abs}$", "$E_{phonon}^{q+1}$", "$E_{em}$", "$E_{phonon}^{q}$"]] for generating example plot.
    example_plot = whether to plot example (True) or override with user values (False).
    \tDefault: True
    example_show_variable_labels = plot A_i_q, A_i_q+1, A_f_q+1 and A_f_q labels on example plot.
    These represent the geometries and charge states of each defect, and relate directly to the formation energies
    that need to be entered into the accepter_to_cbm_calculator() function.
    \tDefault: False
    example_show_process_labels = plot labels for absorption, emission and phonon processes on example plot.
    \tDefault: False
    plot_title = name of output file.
    \tDefault: "optical_transition"
    phonon_labels = plot phonon process labels on non-example plot.
    \tDefault: False (for many transitions, this can clog up the diagram. They also don't position as well
    \tas the absorption and emission labels for diagrams with more than 4 defect geometries.) 
    list_of_plotting_colours = comma separated list of matplotlib colours for plotting diagram. Takes three entries.
    \tFirst determines colour of geometry curves, second controls absorption processes, third controls emission processes.
    \tDefault: ["silver", "red", "blue"].
    convert_abs_em_to_nm = add label to absorption and emission energies showing the wavelength in nm.
    \tDefault: False
    fmt = format of plot. Can be "pdf" or "png".
    \tDefault: "pdf"
    """    
    
    # check inputs
    
    if type(list_of_energies) != list:
        raise TypeError("""Did you enter list_of_energies as a list of lists? This can be a list of the outputs from
        the acceptor_to_cbm_calculator() function.""")
        for i in list_of_energies:
            if type(i) != list:
                raise TypeError("""Did you enter list_of_energies as a list of lists? This can be a list of the outputs from
                the acceptor_to_cbm_calculator() function.""")
    if len(list_of_defects) != len(list_of_energies) + 1:
        raise ValueError("""The number of defect geometries and transition data do not match up. Ensure you have
        n lots of defects and n-1 lots of defect energy entries.""")
    for entry in list_of_defects:
        if entry[0] != "$" or entry[-1] != "$":
            raise ValueError("""It doesn't look like you have entered your defect species in LaTeX math mode format.
            Not doing this will cause scaling problems in the plot. If you have a neutral copper vacancy, you should
            enter this label as '$V_{Cu}^{0}$', ensuring you begin and end the entry with the $ symbol.""")
    if example_plot==True and list_of_defects != ["$A^{q}$", "$A^{q+1}$"]:
        raise ValueError("""You have selected example_plot = True but have also changed the list_of_defects.
            You cannot plot an example plot with non-default list_of_defects. Either change this, or set
            example_plot = False.""")
    if example_plot==True and list_of_energies != [["$E_{abs}$", "$E_{phonon}^{q+1}$", "$E_{em}$", "$E_{phonon}^{q}$"]]:
        raise ValueError("""You have selected example_plot = True but have also changed the list_of_energies.
            You cannot plot an example plot with non-default list_of_energies. Either change this, or set
            example_plot = False.""")
    if fmt != "pdf" and fmt != "png":
        raise ValueError("You have chosen an unsuitable fmt. Choose either 'pdf' or 'png'.")

    
    # configure plot
    
    plt.rcParams.update({"font.size": 20})
    plt.rcParams['axes.facecolor'] = 'white'
    plt.figure(figsize=(10,8))
    plt.xlabel("Configuration Coordinate / Q")
    plt.ylabel("Energy / eV")
    plt.tick_params(
        axis="both",          
        which='both',      
        bottom=False,      
        top=False,
        left=False,
        right=False,
        labelbottom=False,
        labelleft=False)
    

    # generate parabolas sequentially, convert to arrays, plot and annotate
    
    defect_configurations_x = []
    defect_configurations_y = []
    for i in range(0, len(list_of_defects)):
        defect_configurations_x.append(np.linspace(-1,1,5000) + i*0.4)
        defect_configurations_y.append(np.linspace(-1,1,5000)**2 + i*1.2)

    defect_configurations_x = np.asarray(defect_configurations_x)
    defect_configurations_y = np.asarray(defect_configurations_y)
    
        
    for i in range(0, len(list_of_defects)):
        plt.plot(defect_configurations_x[i], defect_configurations_y[i], color = list_of_plotting_colours[0],
                 label='_nolegend_')
        txt = plt.annotate(str(list_of_defects[i]),
                           (defect_configurations_x[i,0]-0.05, defect_configurations_y[i,50]-0.05))
        txt.set_bbox(dict(facecolor='white', alpha=0.5, boxstyle="round"))
    
    # plot transitions sequentially
    
    for i in range(0, len(list_of_energies)):
        
        plt.plot([np.median(defect_configurations_x[i]), np.median(defect_configurations_x[i])],
                    [i*1.2, 1.36 + i*1.2], color = list_of_plotting_colours[1]) # plot absorption
        plt.plot([np.median(defect_configurations_x[i]), np.median(defect_configurations_x[i])+0.4],
                    [1.36 + i*1.2, (i+1)*1.2], color = list_of_plotting_colours[1], linestyle=":") # plot phonon q'
        plt.plot([np.median(defect_configurations_x[i]+0.4), np.median(defect_configurations_x[i]+0.4)],
                    [(i+1)*1.2, (i+1)*1.2 - 1.04], color = list_of_plotting_colours[2]) # plot emission
        plt.plot([np.median(defect_configurations_x[i]+0.4), np.median(defect_configurations_x[i])],
                    [(i+1)*1.2 - 1.04, i*1.2], color = list_of_plotting_colours[2], linestyle=":") # plot phonon q
        plt.plot([np.median(defect_configurations_x[i]), np.median(defect_configurations_x[i]+0.4)],
                    [i*1.2, (i+1)*1.2],
                    "--k") # plot ZPL
            
        # annotate transition sequentially
        
        # configure nm values
        
        abs_nm_value=""
        em_nm_value=""
        if convert_abs_em_to_nm == True:
            abs_nm_value = "\n" + str(round((6.626e-34 * 2.998e17)/(1.602e-19 * list_of_energies[i][0]))) + " nm"
            em_nm_value = "\n" + str(round((6.626e-34 * 2.998e17)/(1.602e-19 * list_of_energies[i][2]))) + " nm"
        
        # configure example annotations
                                     
        if example_plot == True:
            
            # example plot annotate process labels
            
            if example_show_process_labels == True:             
                txt = plt.annotate(str(list_of_energies[i][0]),
                                ((np.median(defect_configurations_x[i]) - 0.6), ((i*1.2)+0.8)),
                                color = list_of_plotting_colours[1]) # annotate absorption
                txt.set_bbox(dict(facecolor='white', alpha=0.7, boxstyle="round", ec=list_of_plotting_colours[1]))
                txt = plt.annotate(str(list_of_energies[i][2]),
                                ((np.median(defect_configurations_x[i]) + 0.5), (i*1.2)+0.8),
                                color = list_of_plotting_colours[2]) # annotate emission
                txt.set_bbox(dict(facecolor='white', alpha=0.7, boxstyle="round", ec=list_of_plotting_colours[2]))
                txt = plt.annotate(str(list_of_energies[i][1]),
                                ((np.median(defect_configurations_x[i])), ((i*1.2)+ 1.5)),
                                color = list_of_plotting_colours[1]) # annotate phonon q'
                txt.set_bbox(dict(facecolor='white', alpha=0.7, boxstyle="round", ec=list_of_plotting_colours[1], ls=":"))
                txt = plt.annotate(str(list_of_energies[i][3]),
                                ((np.median(defect_configurations_x[i]) + 0.4), (i*1.2)),
                                color = list_of_plotting_colours[2]) # annotate phonon q
                txt.set_bbox(dict(facecolor='white', alpha=0.7, boxstyle="round", ec=list_of_plotting_colours[2], ls=":"))
            
            # example plot annotate variable labels
            
            if example_show_variable_labels == True:
                txt = plt.annotate(text="$A_i^q$", xy=(0,0), xytext=(-0.4,0.3), color = "black",
                              arrowprops=dict(arrowstyle="->", connectionstyle="arc3"),
                              fontsize="x-small") # show e_i_q
                txt.set_bbox(dict(facecolor='grey', alpha=0.5, boxstyle="round"))
                txt = plt.annotate(text="$A_i^{q+1}$", xy=(0,1.36), xytext=(-0.6,1.2), color = "black",
                              arrowprops=dict(arrowstyle="->", connectionstyle="arc3"),
                              fontsize="x-small") # show e_i_q+1
                txt.set_bbox(dict(facecolor='grey', alpha=0.5, boxstyle="round"))
                txt = plt.annotate(text="$A_f^{q+1}$", xy=(0.4,1.2), xytext=(0.8,1), color = "black",
                              arrowprops=dict(arrowstyle="->", connectionstyle="arc3"),
                              fontsize="x-small") # show e_f_q+1
                txt.set_bbox(dict(facecolor='grey', alpha=0.5, boxstyle="round"))
                txt = plt.annotate(text="$A_f^{q}$", xy=(0.4,0.16), xytext=(0.9,0.4), color = "black",
                              arrowprops=dict(arrowstyle="->", connectionstyle="arc3"),
                              fontsize="x-small") # show e_f_q
                txt.set_bbox(dict(facecolor='grey', alpha=0.5, boxstyle="round"))
        
        # configure non-example annotations
        
        else:
            txt = plt.annotate(str(round(list_of_energies[i][0], 2)) + " eV" + abs_nm_value,
                            ((np.median(defect_configurations_x[i]) - 0.6), ((i*1.2)+0.7)),
                            color = list_of_plotting_colours[1], fontsize="small") # annotate absorption
            txt.set_bbox(dict(facecolor='white', alpha=0.7, boxstyle="round", ec=list_of_plotting_colours[1]))
            txt = plt.annotate(str(round(list_of_energies[i][2], 2)) + " eV" + em_nm_value,
                            ((np.median(defect_configurations_x[i]) + 0.5), (i*1.2)+0.7),
                            color = list_of_plotting_colours[2], fontsize="small") # annotate emission
            txt.set_bbox(dict(facecolor='white', alpha=0.7, boxstyle="round", ec=list_of_plotting_colours[2]))

            if phonon_labels == True:         
                txt = plt.annotate(str(round(list_of_energies[i][1], 2)) + " eV",
                            ((np.median(defect_configurations_x[i])), ((i*1.2)+ 1.45)),
                            color = list_of_plotting_colours[1], fontsize="xx-small") # annotate phonon q'
                txt.set_bbox(dict(facecolor='white', alpha=0.7, boxstyle="round", ec=list_of_plotting_colours[1], ls=":"))
                txt = plt.annotate(str(round(list_of_energies[i][3], 2)) + " eV",
                            ((np.median(defect_configurations_x[i]) + 0.4), (i*1.2)),
                            color = list_of_plotting_colours[2], fontsize="xx-small") # annotate phonon q
                txt.set_bbox(dict(facecolor='white', alpha=0.7, boxstyle="round", ec=list_of_plotting_colours[2], ls=":"))
    
    # configure legend, save and show
    
    plt.legend(labels = ["$E_{abs}$", "$E_{phonon}^{q+1}$", "$E_{em}$", "$E_{phonon}^{q}$", "ZPL"],
               bbox_to_anchor=(1,1)) # fix legend position
    plt.tight_layout()
    plt.savefig(f"{plot_title}.{fmt}", dpi=600, transparent=False, facecolor="white")
    plt.show()
  
```

```python
cacup_cu_vacancy_example = acceptor_to_cbm_calculator(band_gap = 1.428,
                            q = -1,
                            qplusone= 0,
                            E_i_q = 0.256101,
                            E_i_qplusone = 0.320746,
                            E_f_qplusone = 0.256898,
                            E_f_q= 0.322835)

# how to use the acceptor_to_cbm_calculator() function.
# note that these energies are the formation energies of these respective species
# rather than actual points on the configuration coordinate diagram.
# the pointers enabled with example_show_variable_labels below denote the geometry and charge state
# of the species at each point.

# for a -1 --> 0 transition, you need to calculate defects:
# -1 in -1 geometry, 0 in -1 geometry, 0 in 0 geometry, and -1 in 0 geometry,
# and the formation energies of these correspond to E_i_q, E_i_qplusone, E_f_qplusone, and E_f_q.
```
